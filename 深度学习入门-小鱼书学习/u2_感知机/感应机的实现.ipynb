下面得代码为本人看书之后，结合个人想法，优化后的版本，比如W与b得值，我认为不可能是固定得结果，必然有外部输入

```python
# 与门的实现,W是权重的数组，X是输入的数组，b是偏置
def AND(W,X,b):
    tmp = np.dot(W,X) + b
    if tmp > 0:
        return 1
    else:
        return 0
```

实现与非门，只需要将W值全部取反即可

```python
def NOT_AND(W,X,b):
    tmp=np.dot(-W,X)-b;
    if tmp >= 0:
        return 1
    else:
        return 0
```

```python
def OR(W,X,b):
    tmp=np.dot(W,X)+b;
    if tmp >= 0:
        return 1
    else:
        return 0
```

```python
def XOR(W1,W2,W3,X,b1,b2,b3):
    X1=OR(W1,X,b1)
    X2=NOT_AND(W2,X,b2)
    X3=np.array([X1,X2])
    Y = AND(W3,X3,b3)
    return Y
```
